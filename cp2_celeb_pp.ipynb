```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python

df = pd.read_csv('/content/drive/MyDrive/cp2/fan_RFM_open.csv')
df.loc[137, '메인 플랫폼 링크'] = 'https://www.twitch.tv/luvming01'
```

```python
df.rename(columns = {'아프리카(애청자/팬클럽)' : '아프리카'}, inplace = True)
df.rename(columns = {'유튜브 수' : '유튜브'}, inplace = True)
df.rename(columns = {'메인 평균 방송시간(9개)/업로드주기' : '평균 방송 시간'}, inplace = True)
df.rename(columns = {'평균 시청 횟수/최근 조회수' : '평균 시청 회수'}, inplace = True)
df
```

```python
def find_platform(x):
  if 'twitch' in str(x):
    return 'twitch'
  elif 'youtube' in str(x):
    return 'youtube'
  elif 'spooncast' in str(x):
    return 'spooncast'
  elif 'afreeca' in str(x):
    return 'afreeca'
  elif 'tiktok' in str(x):
    return 'tiktok'
  else:
    return np.NaN
 
  


df['메인 플랫폼'] = df['메인 플랫폼 링크'].apply(find_platform)

list_MainPlatform = []
for i in range(len(df)):
  if str(df['메인 플랫폼'][i]) == 'twitch':
    list_MainPlatform.append(df['트위치'][i])
  elif str(df['메인 플랫폼'][i]) == 'youtube':
    list_MainPlatform.append(df['유튜브'][i])
  elif str(df['메인 플랫폼'][i]) == 'afreeca':
    if '/' in str(df['아프리카'][i]):
      a, b = str(df['아프리카'][i]).split('/')
      list_MainPlatform.append(int(a))
    else:
      list_MainPlatform.append(df['유튜브'][i])
  else:
    list_MainPlatform.append(np.nan)
    
df['메인 플랫폼 구독수'] = list_MainPlatform
```

```python
list_tmp = []
for c in df['콘텐츠 카테고리'].unique():
  if '/' in str(c):
    tmp = c.split('/')
    for x in tmp:
      list_tmp.append(x)
  elif ',' in str(c):
    tmp = c.split(',')
    for x in tmp:
      list_tmp.append(x)
  else:
    list_tmp.append(c)
df_tmp = pd.DataFrame(list_tmp)
list_contents = list(df_tmp[0].unique())
```

```python
for c in list_contents:
  if str(c) == '0' or str(c) in 'nan':
    continue
  print(c)
  c = c.replace(' ', '')

  df[c] = ['X']*len(df)


for i in range(len(df)):
  contents = str(df['콘텐츠 카테고리'][i])
  if ' ' in contents:
    contents = contents.replace(' ', '')
                  
  if '/' in contents:
    tmp = contents.split('/')
    for t in tmp:
      if t in df.columns:
        df[t][i] = 'O'

  elif ',' in contents:
    tmp = contents.split(',')
    for t in tmp:
      if t in df.columns:
        df[t][i] = 'O'
  else:
    if contents in df.columns:
      df[contents][i] = 'O'

df
```

```python
df[df['커뮤니티 링크'] == '0']['커뮤니티 링크']
df['커뮤니티 링크'] = df['커뮤니티 링크'].fillna('X')
df['커뮤니티 링크'].unique()
```

```python
def check_comm(x):
  if x == 'x' or x == 'X' or x =='0':
    return 'X'
  else:
    return 'O'

df['커뮤니티 링크'] = df['커뮤니티 링크'].apply(check_comm)
```

```python
df['커뮤니티 링크'].unique()
```

```python
df.loc[df[df['평균 시청 회수'] == '1만회'].index, '평균 시청 회수'] = '10000회'
df.loc[df[df['평균 시청 회수'] == '8만회'].index, '평균 시청 회수'] = '80000회'

def remove_char(x):
  if type(x) == float:
    isNaN = np.isnan(x)
    if isNaN:
      return np.nan
  x = x.replace('회', '')
  x = x.replace(',', '')
  return int(x)

df['평균 시청 회수'] = df['평균 시청 회수'].apply(remove_char)
df['평균 시청 회수'].unique()

```
