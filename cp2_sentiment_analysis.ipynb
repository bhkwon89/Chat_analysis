```python
df = pd.read_csv('/content/drive/MyDrive/cp2/chat_pp.csv', index_col=0)
```

```python
import re
# ㅋ, ㅎ 제거
# 특수문자 제거

df['message2'] = df['message'].apply(lambda x: re.sub(r'[ㅋㅎ]', '', x))
df['message2'] = df['message'].apply(lambda x: re.sub(r'[^A-Za-z0-9가-힣 ]', '', x))

```

```python
df = df.drop(df[df['message2'].str.contains('참여했습니다')].index)
df = df.drop(df[df['message2'].str.contains('나갔습니다')].index)
df = df.drop(df[df['message2'] == ''].index)
df = df.drop(df[df['message2'] == ' '].index)
df = df.drop(df[df['message2'] == '  '].index)
```

```python
import datetime as dt
```

```python
df['YMDH'] = df['year']*1000000 + df['month']*10000 + df['day']*100 + df['hour']
```

```python

df.to_csv('/content/drive/MyDrive/cp2/chat_pp2.csv')
```

```python
list_ch = df['channel_url'].unique()
df_MessagePerHour = pd.DataFrame()

for i in range(len(list_ch)):
  ch = list_ch[i]
  print(i,'/',len(list_ch))
  tmp_df = df.groupby('channel_url').get_group(ch)

  list_time = tmp_df['YMDH'].unique()
  
  for t in list_time:
    
    time_df = tmp_df.groupby('YMDH').get_group(t)
    
    text = ' '.join(time_df['message2'])
    df_MessagePerHour = df_MessagePerHour.append(pd.DataFrame(
        {
            'channel_url' : [ch],
            'YMDH': [t],
            'text': [text]      
        }
    ))
    
```
